```python
import sqlite3
import json

# SQLite 데이터베이스 연결 (이미 존재하는 데이터베이스 사용)
conn = sqlite3.connect('./db.sqlite3')
cur = conn.cursor()

with open('/Users/youyoungcheon/Desktop/django-project/crawling/remember_job_list.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# 데이터 삽입 함수
def insert_recruit_from_json(json_record):
    cur.execute('''INSERT INTO recruits_recruit 
                   (title, company_name, detail_url, end_date, platform_name, category_name, stack, region, career) 
                   VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)''',
                (json_record['title'], json_record['company_name'], json_record['detail_url'], 
                 json_record['end_date'], json_record['platform_name'], json_record['category_name'], 
                 json_record['stack'], json_record['region'], json_record['career']))
    conn.commit()

# JSON 데이터를 SQLite에 삽입
for record in data:
    insert_recruit_from_json(record)

# 데이터 확인을 위해 조회
cur.execute("SELECT * FROM recruits_recruit")
rows = cur.fetchall()
for row in rows:
    print(row)

# 데이터베이스 연결 종료
conn.close()
```
